```python
import os
import requests
import zipfile

def download_and_extract_files(zones, save_dir):
    # Base URL e sufixos para cada tipo de arquivo
    base_url = 'http://www.dsr.inpe.br/topodata/data/geotiff/'
    file_suffixes = ['ZN', 'SN', 'ON', 'RS', 'FT', 'DD', 'VN', 'HN']

    # Certifique-se de que o diretório de destino existe
    os.makedirs(save_dir, exist_ok=True)

    # Função para baixar e extrair um arquivo
    def download_and_extract(url, save_path):
        response = requests.get(url)
        if response.status_code == 200:
            zip_path = save_path + '.zip'
            with open(zip_path, 'wb') as f:
                f.write(response.content)
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(save_path)
            os.remove(zip_path)  # Remove o arquivo zip após a extração
            print(f"Extraído: {save_path}")
        else:
            print(f"Erro ao baixar o arquivo {url}")

    # Processar cada zona e cada tipo de arquivo
    for zone in zones:
        for suffix in file_suffixes:
            file_code = f"{zone}{suffix}"
            url = f"{base_url}{file_code}.zip"
            save_path = os.path.join(save_dir, file_code)
            print(url)
            download_and_extract(url, save_path)

# Lista de zonas
zones = [
    '29S54_', '30S54_', '31S54_', '32S54_', '33S54_',
    '29S525', '30S525', '31S525', '32S525',
    '29S51_', '30S51_', '31S51_'
]

# Diretório para salvar os arquivos
save_dir = 'zonas'  # Substitua pelo seu diretório de destino

download_and_extract_files(zones, save_dir)
```
